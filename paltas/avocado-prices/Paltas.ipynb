```python
import pandas as pd
```

```python
paltas = pd.read_csv("avocado.csv")
paltas["Date"]
parametros = ["Date","AveragePrice","Total Volume","Total Bags","Small Bags","Large Bags","XLarge Bags","type","region"]
paltas = paltas[parametros]
paltas.head()
```

```python
paltas.groupby(["region","type"])["AveragePrice"].agg([min,max,sum]).head()
```

```python
def pct5(column):
    return column.quantile(0.05)

def pct95(column):
    return column.quantile(0.95)
```

```python
paltas[paltas["region"]=="Boston"]["AveragePrice"].median()
```

```python
paltas[paltas["region"]=="Boston"]["AveragePrice"].agg([pct5,pct95])
```
